```python
import pandas as pd
import os
```

```python
CSV_PATH = os.path.abspath('transcribes/New channel meeting-20230706_092135-Meeting Recording.mp4.csv')

file_name, _ = os.path.splitext(os.path.basename(CSV_PATH))

TEXT_PATH = os.path.join('transcribes', file_name + '.txt')
```

```python
df = pd.read_csv(CSV_PATH, index_col='id')
```

```python
text = ' '.join(df['text'])
```

```python
text
```

```python
with open(TEXT_PATH, "w", encoding="utf-8") as f:
    f.write(text)
```
