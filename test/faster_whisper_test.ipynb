```python
import os.path
from torch import cuda
import pandas as pd
import tqdm
import itertools

SD_PATH = os.path.abspath('videos\New channel meeting-20230706_092135-Meeting Recording.mp4')

DEVICE = "cuda" if cuda.is_available() else "cpu"
```

```python
from faster_whisper import WhisperModel

model_size = "large-v2"  # tiny base small medium large-v2

# Run on GPU with FP16
model = WhisperModel(model_size, device=DEVICE, compute_type="float16")
```

```python
file_name = os.path.basename(SD_PATH)

segments, info = model.transcribe(SD_PATH, beam_size=3, vad_filter=True)

print("Detected language '%s' with probability %f" % (info.language, info.language_probability))

transcribe_df = pd.DataFrame()
```

```python
for t in segments:
    d = {
        'id': t.id,
        'seek': t.seek,
        'start': t.start,
        'end': t.end,
        'text': t.text,
        'temperature': t.temperature,
        'avg_logprob': t.avg_logprob,
        'compression_ratio': t.compression_ratio,
        'no_speech_prob': t.no_speech_prob,
        'words': t.words,
    }
    transcribe_df = pd.concat([transcribe_df, pd.DataFrame([d])], ignore_index=True)
    transcribe_df.to_csv(os.path.join('transcribes', file_name + '.csv'))
```
